### For Machine Learning (LossFunction= mean Square Error)

```python
import numpy as np    
import pandas as pd
```

```python
def gredient_descent(x,y):
    m = c = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.06
    for i in range(iterations):
        k = []
        y_pred = m*x+c
        cost = (1/n)*sum([val**2 for val in (y-y_pred)])
        md = -(2/n)*sum(x*(y-y_pred))
        cd = -(2/n)*sum(y-y_pred)
        m = m - learning_rate * md
        c = c - learning_rate * cd
        k.append({'m_value':m,'c_value':c,'cost':cost,'iteration':i})
        print(k)
#         print("m{},c{},cost{} for iteration {}".format(m,c,cost,i))
x = np.array([1,2,3,4,5,6])
y = np.array([5,6,7,2,4,5])


gredient_descent(x,y)
```

### Formulas:

- y_pred = m*x+c


#### cost is given by

- cost = (1/n)*sum( (y-y_pred)** 2)

#### partial derivatives of slop(m) and c formula 
- md = -(2/n)*sum(x*(y-y_pred))
- cd = -(2/n)*sum(y-y_pred)

#### Slop and consatant update formula with learning rate

- m = m - learning_rate * md
- c = c - learning_rate * cd
